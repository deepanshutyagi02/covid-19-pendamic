```python
import pandas as pd

try:
    df = pd.read_csv('covid_19_clean_complete.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'covid_19_clean_complete.csv' not found.")
except Exception as e:
    print(f"An error occurred: {e}")
```

```python
# Display DataFrame info
print(df.info())

# Identify key variables and check date format
print("\nKey variables:")
print(df[['Date', 'Country/Region', 'Confirmed', 'Deaths', 'Recovered']].head())
print("\nDate column data type:", df['Date'].dtype)

# Check for missing values
print("\nMissing values per column:")
print(df.isnull().sum())

# Check for inconsistencies in categorical columns
print("\nUnique values in 'Country/Region':")
print(df['Country/Region'].unique())
print("\nUnique values in 'WHO Region':")
print(df['WHO Region'].unique())

# Check for inconsistencies in numerical columns
print("\nDescriptive statistics for numerical columns:")
print(df[['Confirmed', 'Deaths', 'Recovered']].describe())

# Check for negative values in numerical columns
print("\nNumber of negative values in 'Confirmed':", (df['Confirmed'] < 0).sum())
print("Number of negative values in 'Deaths':", (df['Deaths'] < 0).sum())
print("Number of negative values in 'Recovered':", (df['Recovered'] < 0).sum())
```

```python
# Fill missing 'Province/State' with 'Unknown'
df['Province/State'].fillna('Unknown', inplace=True)

# Convert 'Date' to datetime objects
try:
    df['Date'] = pd.to_datetime(df['Date'])
except ValueError:
    # Handle potential errors during conversion (e.g., incorrect date format)
    print("Error converting 'Date' to datetime. Inspect the 'Date' column for unusual values.")
    # Example of how to handle errors with a specific date format:
    # df['Date'] = pd.to_datetime(df['Date'], format='%m/%d/%y')

# Check and fill missing values in key numerical columns
for col in ['Confirmed', 'Deaths', 'Recovered']:
    if df[col].isnull().any():
        print(f"Filling missing values in '{col}' with 0")
        df[col].fillna(0, inplace=True)

# Check for negative values in numerical columns and handle appropriately
for col in ['Confirmed', 'Deaths', 'Recovered']:
  if (df[col] < 0).any():
    print(f"Negative values found in '{col}'. Replacing them with 0.")
    df[col] = df[col].clip(lower=0)

# Display the updated DataFrame info for verification.
print(df.info())
display(df.head())
```

```python
import numpy as np

# Calculate mortality and recovery rates
df['Mortality Rate'] = np.where(df['Confirmed'] > 0, df['Deaths'] / df['Confirmed'], 0)
df['Recovery Rate'] = np.where(df['Confirmed'] > 0, df['Recovered'] / df['Confirmed'], 0)

# Calculate daily changes
df['Daily New Cases'] = df.groupby('Country/Region')['Confirmed'].diff().fillna(0)
df['Daily New Deaths'] = df.groupby('Country/Region')['Deaths'].diff().fillna(0)
df['Daily New Recovered'] = df.groupby('Country/Region')['Recovered'].diff().fillna(0)

# Aggregate data
df_aggregated = df.groupby(['Country/Region', 'Date']).agg({
    'Confirmed': 'sum',
    'Deaths': 'sum',
    'Recovered': 'sum',
    'Mortality Rate': 'mean',  # Use mean for rates
    'Recovery Rate': 'mean',
    'Daily New Cases': 'sum',
    'Daily New Deaths': 'sum',
    'Daily New Recovered': 'sum',
}).reset_index()

# Ensure correct data types
for col in ['Confirmed', 'Deaths', 'Recovered', 'Daily New Cases', 'Daily New Deaths', 'Daily New Recovered']:
    df_aggregated[col] = df_aggregated[col].astype('int64')
df_aggregated['Date'] = pd.to_datetime(df_aggregated['Date'])
display(df_aggregated.head())
print(df_aggregated.info())
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Overall Trends
total_cases_by_country = df_aggregated.groupby('Country/Region')['Confirmed'].max()
total_deaths_by_country = df_aggregated.groupby('Country/Region')['Deaths'].max()
avg_daily_new_cases = df_aggregated.groupby('Country/Region')['Daily New Cases'].mean()

# Plotting overall trends (example: total cases)
plt.figure(figsize=(10, 6))
total_cases_by_country.sort_values(ascending=False).head(10).plot(kind='bar', color='skyblue')
plt.title('Top 10 Countries with Highest Total Confirmed Cases')
plt.xlabel('Country/Region')
plt.ylabel('Total Confirmed Cases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 2. Distribution across countries
plt.figure(figsize=(12, 6))
sns.histplot(total_cases_by_country, kde=True, color='salmon')
plt.title('Distribution of Total Confirmed Cases Across Countries')
plt.xlabel('Total Confirmed Cases')
plt.ylabel('Number of Countries')
plt.show()

# Correlation analysis
correlation_matrix = df_aggregated[['Confirmed', 'Deaths', 'Recovered']].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='viridis', fmt=".2f")
plt.title('Correlation Matrix of Confirmed Cases, Deaths, and Recoveries')
plt.show()

# 3. Relationships between engineered features
# Example: Scatter plot of mortality rate vs. confirmed cases
plt.figure(figsize=(10, 6))
sns.scatterplot(x='Confirmed', y='Mortality Rate', data=df_aggregated, hue='Country/Region', palette='magma', legend=False)  # Remove legend for better visibility
plt.title('Mortality Rate vs. Confirmed Cases')
plt.xlabel('Confirmed Cases')
plt.ylabel('Mortality Rate')
plt.show()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 3. Relationships between engineered features (continued)
# Example: Line plot of daily new cases over time for a specific country
country = 'US'  # Replace with desired country
country_data = df_aggregated[df_aggregated['Country/Region'] == country]
plt.figure(figsize=(12, 6))
plt.plot(country_data['Date'], country_data['Daily New Cases'], color='green', label='Daily New Cases')
plt.xlabel('Date')
plt.ylabel('Daily New Cases')
plt.title(f'Daily New Cases in {country} Over Time')
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()

# 4. Data Segmentation
# Example: Compare trends before and after a specific date
date_threshold = '2020-03-15'  # Replace with desired date

df_before = df_aggregated[df_aggregated['Date'] < date_threshold]
df_after = df_aggregated[df_aggregated['Date'] >= date_threshold]

plt.figure(figsize=(12, 6))
sns.lineplot(x='Date', y='Daily New Cases', data=df_before, label='Before ' + date_threshold, color='blue')
sns.lineplot(x='Date', y='Daily New Cases', data=df_after, label='After ' + date_threshold, color='red')
plt.xlabel('Date')
plt.ylabel('Daily New Cases')
plt.title('Daily New Cases Before and After ' + date_threshold)
plt.xticks(rotation=45, ha='right')
plt.legend()
plt.tight_layout()
plt.show()
```
