```
import pandas as pd

try:
    df = pd.read_csv('country_wise_latest.csv')
    display(df.head())
    print(df.shape)
except FileNotFoundError:
    print("Error: 'country_wise_latest.csv' not found. Please ensure the file exists in the current directory.")
    df = None
except pd.errors.ParserError:
    print("Error: Unable to parse the CSV file. Please check the file format.")
    df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

```
# Display DataFrame information
df.info()

# Generate descriptive statistics
display(df.describe())

# Examine unique values in 'Country/Region' and 'WHO Region' columns
print(f"\nUnique Countries/Regions:\n{df['Country/Region'].unique()}")
print(f"\nUnique WHO Regions:\n{df['WHO Region'].unique()}")

# Check for missing values
print(f"\nMissing values per column:\n{df.isnull().sum()}")
```

```
# Convert 'Confirmed', 'Deaths', and 'Recovered' columns to numeric, coercing errors to NaN
for col in ['Confirmed', 'Deaths', 'Recovered']:
    df[col] = pd.to_numeric(df[col], errors='coerce')

# Check for and handle inconsistencies in 'WHO Region'
# (Example: Replace typos, standardize names)
df['WHO Region'] = df['WHO Region'].astype('category')
# (Add more specific cleaning based on observed inconsistencies, if any)

# Review 'Country/Region' for inconsistencies and duplicates
# (Example: Correct typos, handle duplicates by aggregation or removal)
# (Add more specific cleaning based on observed inconsistencies and duplicates, if any)

# Verify data types and missing values after cleaning
print(df.info())
print(df.isnull().sum())

display(df.head())
```

```
# Calculate descriptive statistics
numerical_cols = ['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths', 'New recovered']
descriptive_stats = df[numerical_cols].describe()
display(descriptive_stats)

# Analyze the relationship between 'Confirmed' cases and 'Deaths'
correlation_confirmed_deaths = df['Confirmed'].corr(df['Deaths'])
print(f"\nCorrelation between Confirmed cases and Deaths: {correlation_confirmed_deaths}")

# Group data by 'WHO Region'
grouped_by_region = df.groupby('WHO Region')[['Confirmed', 'Deaths', 'Recovered']].sum()
display(grouped_by_region)

# Identify potential outliers using box plots
import matplotlib.pyplot as plt
plt.figure(figsize=(15, 10))
df[numerical_cols].boxplot()
plt.xticks(rotation=45, ha='right')
plt.title('Box Plots of Numerical Columns')
plt.show()
```

```
import matplotlib.pyplot as plt

# 1. Top 10 countries with the highest confirmed cases
top_10_countries = df.nlargest(10, 'Confirmed')
plt.figure(figsize=(12, 6))
plt.bar(top_10_countries['Country/Region'], top_10_countries['Confirmed'], color='skyblue')
plt.xlabel('Country/Region')
plt.ylabel('Confirmed Cases')
plt.title('Top 10 Countries with Highest Confirmed Cases')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()


# 2. Total confirmed cases, deaths, and recoveries per WHO region
region_stats = df.groupby('WHO Region')[['Confirmed', 'Deaths', 'Recovered']].sum()
region_stats.plot(kind='bar', figsize=(12, 6), color=['skyblue', 'salmon', 'lightgreen'])
plt.xlabel('WHO Region')
plt.ylabel('Number of Cases')
plt.title('Total Confirmed Cases, Deaths, and Recoveries per WHO Region')
plt.xticks(rotation=45, ha='right')
plt.legend(title='Case Type')
plt.tight_layout()
plt.show()


# 3. Scatter plot: Confirmed cases vs. Deaths, colored by WHO region
plt.figure(figsize=(10, 6))
for region in df['WHO Region'].unique():
    subset = df[df['WHO Region'] == region]
    plt.scatter(subset['Confirmed'], subset['Deaths'], label=region, alpha=0.7)
plt.xlabel('Confirmed Cases')
plt.ylabel('Deaths')
plt.title('Confirmed Cases vs. Deaths (Colored by WHO Region)')
plt.legend(title='WHO Region', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()


# 4. Line plot: Trend of new cases over time (if time-series data is available)
# Since the provided dataset does not include a date or time column, a line plot cannot be created.
print("No time-series data available to plot the trend of new cases over time.")
```
