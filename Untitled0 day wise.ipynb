## Data loading

### Subtask:
Load the "day_wise.csv" file into a pandas DataFrame.


**Reasoning**:
Load the "day_wise.csv" file into a pandas DataFrame and display the first 5 rows.



```python
import pandas as pd

try:
    df = pd.read_csv('day_wise.csv')
    display(df.head())
except FileNotFoundError:
    print("Error: 'day_wise.csv' not found. Please ensure the file exists in the current directory or provide the correct path.")
except pd.errors.ParserError:
    print("Error: Could not parse 'day_wise.csv'. Please check the file format.")
except Exception as e:
    print(f"An unexpected error occurred: {e}")
```

## Data exploration

### Subtask:
Explore the loaded data to understand its structure, identify data types of columns, check for missing values, and examine the distribution of key variables.


**Reasoning**:
Explore the data shape, info, descriptive statistics, missing values, data types, and distributions of key variables as instructed.



```python
# Data Shape and Info
print("Data Shape:", df.shape)
display(df.info())

# Descriptive Statistics
display(df.describe())

# Missing Values Analysis
print("\nMissing Values per Column:\n", df.isnull().sum())
import matplotlib.pyplot as plt
import seaborn as sns
plt.figure(figsize=(10, 6))
sns.heatmap(df.isnull(), cbar=False)
plt.title("Missing Values Heatmap")
plt.show()


# Data Type Exploration
print("\nData Types:\n", df.dtypes)
# Convert 'Date' column to datetime if necessary
if not pd.api.types.is_datetime64_any_dtype(df['Date']):
    df['Date'] = pd.to_datetime(df['Date'])
print("\nData Types after conversion:\n", df.dtypes)

# Distribution of Key Variables
key_variables = ['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths', 'New recovered']
plt.figure(figsize=(15, 10))
for i, var in enumerate(key_variables):
    plt.subplot(3, 3, i + 1)
    sns.histplot(df[var], kde=True)
    plt.title(f'Distribution of {var}')
plt.tight_layout()
plt.show()

plt.figure(figsize=(15, 10))
for i, var in enumerate(key_variables):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df[var])
    plt.title(f'Boxplot of {var}')
plt.tight_layout()
plt.show()

# Correlation Analysis
correlation_matrix = df.corr()
plt.figure(figsize=(12, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Features')
plt.show()
```

## Data cleaning

### Subtask:
Clean the data by handling inconsistencies and missing values.


**Reasoning**:
Check for inconsistencies like negative values in 'New cases', 'New deaths', and 'New recovered', handle missing values (though none were found in the exploration step), verify data types, and check for outliers in numerical columns using boxplots.



```python
# Check for inconsistencies (negative values)
for col in ['New cases', 'New deaths', 'New recovered']:
    if (df[col] < 0).any():
        print(f"Warning: Negative values found in '{col}' column.")
        # Handle negative values (e.g., replace with 0 or the mean)
        df[col] = df[col].clip(lower=0)  # Replace negative values with 0


# Verify data types (already checked in previous step, but double-checking)
for col in ['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths', 'New recovered']:
    if not pd.api.types.is_numeric_dtype(df[col]):
        print(f"Warning: '{col}' column is not numeric. Attempting conversion...")
        try:
            df[col] = pd.to_numeric(df[col], errors='coerce')
        except Exception as e:
            print(f"Error converting '{col}' to numeric: {e}")

# Check for and handle outliers (using boxplots for visualization and IQR for outlier detection)
import matplotlib.pyplot as plt
import seaborn as sns

numerical_cols = ['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths', 'New recovered']
plt.figure(figsize=(15, 10))
for i, col in enumerate(numerical_cols):
    plt.subplot(3, 3, i + 1)
    sns.boxplot(y=df[col])
    plt.title(f'Boxplot of {col}')

    # Calculate IQR for outlier detection
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    # Identify outliers
    outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
    print(f"Outliers in {col}: \n{outliers}")

plt.tight_layout()
plt.show()
display(df.describe())
```

## Data analysis

### Subtask:
Analyze the cleaned data to identify trends, patterns, and relationships between variables.


**Reasoning**:
Calculate descriptive statistics, explore relationships between variables, analyze trends over time, and investigate the relationships between new and cumulative cases.



```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Descriptive Statistics
numerical_cols = ['Confirmed', 'Deaths', 'Recovered', 'Active', 'New cases', 'New deaths', 'New recovered']
display(df[numerical_cols].describe())

# 2. Relationships between variables (Correlation and Scatter Plots)
correlation_matrix = df[['Confirmed', 'Deaths', 'Recovered', 'Active']].corr()
display(correlation_matrix)
plt.figure(figsize=(10, 8))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix')
plt.show()

plt.figure(figsize=(15, 5))
for i, col in enumerate(['Deaths', 'Recovered', 'Active']):
    plt.subplot(1, 3, i + 1)
    sns.regplot(x='Confirmed', y=col, data=df)
    plt.title(f'Confirmed vs. {col}')
plt.tight_layout()
plt.show()


# 3. Trends over time (Line Plots with Rolling Averages)
plt.figure(figsize=(15, 10))
for i, col in enumerate(['Confirmed', 'Deaths', 'Recovered', 'Active']):
    plt.subplot(2, 2, i + 1)
    plt.plot(df['Date'], df[col], label=col)
    plt.plot(df['Date'], df[col].rolling(window=7).mean(), label=f'{col} (7-day rolling avg)') # 7-day rolling average
    plt.xlabel('Date')
    plt.ylabel(col)
    plt.title(f'Trend of {col} over time')
    plt.xticks(rotation=45)
    plt.legend()
plt.tight_layout()
plt.show()


# 4. New vs. Cumulative Cases
plt.figure(figsize=(15, 10))

for i, (new_col, cum_col) in enumerate(zip(['New cases', 'New deaths', 'New recovered'], ['Confirmed', 'Deaths', 'Recovered'])):
  plt.subplot(3, 1, i + 1)
  plt.bar(df['Date'], df[new_col], label=new_col, color='skyblue')
  plt.plot(df['Date'], df[cum_col], label=cum_col, color='red')
  plt.xlabel('Date')
  plt.title(f'{new_col} vs {cum_col}')
  plt.xticks(rotation=45)
  plt.legend()

plt.tight_layout()
plt.show()
```
