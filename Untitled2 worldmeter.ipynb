```python
import pandas as pd

try:
    df = pd.read_csv('worldometer_data.csv')
    display(df.head())
except FileNotFoundError:
    print("Error: 'worldometer_data.csv' not found.")
    df = None # Ensure df is defined even in case of error
except pd.errors.ParserError:
    print("Error: Could not parse 'worldometer_data.csv'. Check file format.")
    df = None
except Exception as e:
    print(f"An unexpected error occurred: {e}")
    df = None
```

```python
# Check the shape of the DataFrame
print("Shape of the DataFrame:", df.shape)

# Get a summary of the DataFrame
print("\nInfo:")
df.info()

# Calculate descriptive statistics
print("\nDescriptive Statistics:")
print(df.describe())

# Explore unique values in categorical columns
print("\nUnique Continents:", df['Continent'].unique())
print("\nValue counts for Continent:")
print(df['Continent'].value_counts())
print("\nUnique Countries/Regions:", df['Country/Region'].unique())

# Visualize distributions of key numerical variables
import matplotlib.pyplot as plt

plt.figure(figsize=(15, 5))

plt.subplot(1, 3, 1)
plt.hist(df['TotalCases'].dropna(), bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Total Cases')
plt.xlabel('Total Cases')
plt.ylabel('Frequency')


plt.subplot(1, 3, 2)
plt.hist(df['TotalDeaths'].dropna(), bins=20, color='salmon', edgecolor='black')
plt.title('Distribution of Total Deaths')
plt.xlabel('Total Deaths')
plt.ylabel('Frequency')

plt.subplot(1, 3, 3)
plt.hist(df['TotalRecovered'].dropna(), bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Total Recovered')
plt.xlabel('Total Recovered')
plt.ylabel('Frequency')

plt.tight_layout()
plt.show()
```

```python
# Impute missing numerical values with the mean.
for col in ['TotalDeaths', 'TotalRecovered', 'ActiveCases', 'TotalTests', 'Population']:
    if pd.api.types.is_numeric_dtype(df[col]):
        df[col] = df[col].fillna(df[col].mean())
    else:
        # Attempt to convert to numeric, then fill with mean.
        try:
          df[col] = pd.to_numeric(df[col], errors='coerce')
          df[col] = df[col].fillna(df[col].mean())
        except:
          print(f"Column {col} could not be converted to numeric and imputed.")

# Impute missing categorical values with the mode
for col in ['Continent', 'WHO Region']:
    df[col] = df[col].fillna(df[col].mode()[0])

# Convert relevant columns to appropriate data types
for col in ['TotalCases', 'TotalDeaths', 'TotalRecovered', 'ActiveCases', 'Population', 'TotalTests']:
    try:
        df[col] = pd.to_numeric(df[col], errors='coerce')
    except Exception as e:
        print(f"Error converting column '{col}' to numeric: {e}")

# Remove duplicate rows
num_duplicates = df.duplicated().sum()
df = df.drop_duplicates()
print(f"Removed {num_duplicates} duplicate rows.")

# Display the first few rows of the cleaned DataFrame and its info
display(df.head())
df.info()
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# 1. Descriptive Statistics
numerical_cols = ['TotalCases', 'TotalDeaths', 'TotalRecovered', 'Population']
descriptive_stats = df[numerical_cols].describe()
display(descriptive_stats)

# 2. Correlation Matrix and Heatmap
correlation_matrix = df[numerical_cols].corr()
plt.figure(figsize=(8, 6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Variables')
plt.show()


# 3. Mortality and Recovery Rates
df['MortalityRate'] = df['TotalDeaths'] / df['TotalCases'].replace(0, 1)  # Handle division by zero
df['RecoveryRate'] = df['TotalRecovered'] / df['TotalCases'].replace(0, 1)  # Handle division by zero

plt.figure(figsize=(12, 5))

plt.subplot(1, 2, 1)
plt.hist(df['MortalityRate'], bins=20, color='skyblue', edgecolor='black')
plt.title('Distribution of Mortality Rate')
plt.xlabel('Mortality Rate')
plt.ylabel('Frequency')

plt.subplot(1, 2, 2)
plt.hist(df['RecoveryRate'], bins=20, color='lightgreen', edgecolor='black')
plt.title('Distribution of Recovery Rate')
plt.xlabel('Recovery Rate')
plt.ylabel('Frequency')
plt.tight_layout()
plt.show()

# Grouped analysis by continent
mortality_by_continent = df.groupby('Continent')['MortalityRate'].mean()
recovery_by_continent = df.groupby('Continent')['RecoveryRate'].mean()

display(mortality_by_continent)
display(recovery_by_continent)

# 4. Additional Analysis (Outlier Detection)
plt.figure(figsize=(15, 5))
plt.subplot(1, 3, 1)
sns.boxplot(y=df['TotalCases'], color='skyblue')
plt.title('Total Cases Boxplot')

plt.subplot(1, 3, 2)
sns.boxplot(y=df['TotalDeaths'], color='salmon')
plt.title('Total Deaths Boxplot')

plt.subplot(1, 3, 3)
sns.boxplot(y=df['TotalRecovered'], color='lightgreen')
plt.title('Total Recovered Boxplot')
plt.tight_layout()
plt.show()
```
